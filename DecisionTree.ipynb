```python
import numpy as np
import pandas as pd
```

**Loading Dataset**

```python
df = pd.read_csv('Loan Elgible status.csv')
df
```

**X and y**

```python
X = df.drop(columns=['Loan_Status'])
y= df['Loan_Status']
```

**TrainTestSplit**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

# DecisionTreeClassifier

**Hyperparameter Tuning**

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import GridSearchCV
estimator = DecisionTreeClassifier()
peram_grid = {'criterion':['gini','entropy'],'max_depth':list(range(1,16))}
dt= GridSearchCV(estimator,peram_grid,scoring='accuracy',cv=5)
dt.fit(X_train,y_train)
dt.best_estimator_
```

```python
#Modelling on Traindata
dt  = DecisionTreeClassifier(max_depth=1)
dt.fit(X_train,y_train)

#prediction on Traindata
ypred_train = dt.predict(X_train)

#evaluation
from sklearn.metrics import accuracy_score
print('Train Accurary:',accuracy_score(y_train,ypred_train))
from sklearn.model_selection import cross_val_score
print('Cv_score:',cross_val_score(dt,X_train,y_train,cv=5,scoring='accuracy').mean())

#test dataset
#prediction
y_pred_test = dt.predict(X_test)

#evaluation
from sklearn.metrics import accuracy_score
print('Train Accurary:',accuracy_score(y_test,y_pred_test))
```
